### CLEAN TYPE 1 COLUMNS (YES / NO / NOT DISCUSSED / AMBIGUOUS)

```python
type1_cols = [
    "Serious_Complaint",
    "Other_Complaint",
    "Discussion_on_Price_Increase",
    "Renewal_Impact_Due_to_Price_Increase",
    "Discount_or_Waiver_Requested",
    "Call_Reschedule_Request",
    "Agent_Flagged_Membership_Status_Alert",
    "Agent_Renewal_Initiation",
    "Explicit_Switching_Intent",
    "Price_Switching_Mentioned",
    "Percentage_Price_Increase_Mentioned",
    "Monetary_Price_Increase_Mentioned",
    "Customer_Asked_For_Justification",
    "Discount_Offered",
    "Membership_Renewal_Decision",
]
 
for col in type1_cols:
    if col not in df.columns:
        continue
 
    # Strip formatting characters
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].str.replace(r"\*+", "", regex=True)
    df[col] = df[col].str.replace(r"[\[\]]", "", regex=True)
    df[col] = df[col].str.strip()
 
    # Ambiguous — model hedged between yes and no
    ambiguous_mask = (
        df[col].str.lower().str.contains("yes/no", na=False) |
        df[col].str.lower().str.contains("yes / no", na=False)
    )
    df.loc[ambiguous_mask, col] = "Ambiguous"
 
    # Not Applicable
    na_mask = df[col].str.lower().str.startswith("not applicable", na=False)
    df.loc[na_mask & ~ambiguous_mask, col] = "Not Applicable"
 
    # Not Discussed
    nd_mask = df[col].str.lower().str.startswith("not discussed", na=False)
    df.loc[nd_mask & ~ambiguous_mask, col] = "Not Discussed"
 
    # No — must come before Yes to avoid "No" catching "Not Discussed"
    no_mask = (
        df[col].str.lower().str.startswith("no", na=False) &
        ~nd_mask &
        ~na_mask &
        ~ambiguous_mask
    )
    df.loc[no_mask, col] = "No"
 
    # Yes
    yes_mask = (
        df[col].str.lower().str.startswith("yes", na=False) &
        ~ambiguous_mask
    )
    df.loc[yes_mask, col] = "Yes"
 
    # Fill NaN and unresolved values
    df[col] = df[col].replace("nan", "Not Discussed")
    df[col] = df[col].fillna("Not Discussed")
 
    valid_values = ["Yes", "No", "Not Discussed", "Not Applicable", "Ambiguous"]
    df.loc[~df[col].isin(valid_values), col] = "Ambiguous"
```

#### CLEAN DESIRE_TO_CANCEL

```python
df["Desire_To_Cancel"] = df["Desire_To_Cancel"].astype(str).str.strip()
 
cancel_keywords   = ["cancel", "leave", "closure", "not renew", "liquidation",
                     "closing down", "non-renewal", "termination", "withdrawal",
                     "desired to cancel", "want to cancel"]
 
renew_keywords    = ["renewed", "agreed to renew", "willing to continue",
                     "desired to renew", "renew/not cancelled"]
 
undecided_keywords= ["thinking", "consider", "discuss with", "undecided",
                     "will check", "hesitat", "unclear", "unsure",
                     "renewal undecided", "renewal not discussed", "uncertain"]
 
nd_keywords       = ["not discussed", "not mentioned", "no conversation", "no discussion"]
 
na_keywords       = ["disconnected", "incomplete", "no response",
                     "not initiated", "not applicable"]
 
cancel_mask    = df["Desire_To_Cancel"].str.lower().apply(lambda v: any(k in v for k in cancel_keywords))
renew_mask     = df["Desire_To_Cancel"].str.lower().apply(lambda v: any(k in v for k in renew_keywords))    & ~cancel_mask
undecided_mask = df["Desire_To_Cancel"].str.lower().apply(lambda v: any(k in v for k in undecided_keywords)) & ~cancel_mask & ~renew_mask
nd_mask        = df["Desire_To_Cancel"].str.lower().apply(lambda v: any(k in v for k in nd_keywords))        & ~cancel_mask & ~renew_mask & ~undecided_mask
na_mask        = df["Desire_To_Cancel"].str.lower().apply(lambda v: any(k in v for k in na_keywords))        & ~cancel_mask & ~renew_mask & ~undecided_mask & ~nd_mask
 
df["Desire_To_Cancel"] = "Ambiguous"
df.loc[na_mask,        "Desire_To_Cancel"] = "Not Applicable"
df.loc[nd_mask,        "Desire_To_Cancel"] = "Not Discussed"
df.loc[undecided_mask, "Desire_To_Cancel"] = "Undecided"
df.loc[renew_mask,     "Desire_To_Cancel"] = "Renewed"
df.loc[cancel_mask,    "Desire_To_Cancel"] = "Desired to Cancel"
```

####  CLEAN COMPETITOR COLUMNS

```python
# Common noise values
noise_values = ["xxxx", "unavailable", "prompt 2:", ":", "", "nan"]

if "Explicit_Competitor_Mention" in df.columns:
    
    df["Explicit_Competitor_Mention"] = (
        df["Explicit_Competitor_Mention"]
        .astype(str)
        .str.lower()
        .str.strip()
        .str.replace(r"[\[\]]", "", regex=True)   # handle [No]
    )
    
    # Replace noise → NaN
    df["Explicit_Competitor_Mention"] = df["Explicit_Competitor_Mention"].replace(noise_values, np.nan)

    # Convert to lowercase safely
    col = df["Explicit_Competitor_Mention"].astype(str).str.lower().str.strip()

    # Final binary classification
    df["Explicit_Competitor_Mention"] = np.where(
        col.str.startswith("yes"),
        "Yes",
        "No"
    )


if "Mentioned_Competitors" in df.columns:
    
    df["Mentioned_Competitors"] = (
        df["Mentioned_Competitors"]
        .astype(str)
        .str.lower()
        .str.strip()
        .str.replace(r"[\[\]]", "", regex=True)   # handle [No]
    )
    
   # Replace noise → NaN
    df["Mentioned_Competitors"] = df["Mentioned_Competitors"].replace(noise_values, np.nan)
    
    # Final classification (binary)
    df["Mentioned_Competitors"] = np.where(
        df["Mentioned_Competitors"].isna(), "No",
        np.where(
            df["Mentioned_Competitors"].isin(["no", "not mentioned", "not discussed"]),
            "No",
            "Yes"
        )
    )
```

#### CLEAN ACCORDING TO VALUES

```python
if "Topic_Introduced_By" in df.columns:
    customer_mask = df["Topic_Introduced_By"].astype(str).str.lower().str.contains("customer", na=False)
    agent_mask    = df["Topic_Introduced_By"].astype(str).str.lower().str.contains("agent", na=False)
 
    df["Topic_Introduced_By"] = "Not Discussed"
    df.loc[agent_mask,    "Topic_Introduced_By"] = "Agent"
    df.loc[customer_mask, "Topic_Introduced_By"] = "Customer"
```

```python
if "Customer_Response" in df.columns:
    negative_mask = df["Customer_Response"].astype(str).str.lower().str.contains(
        "negative|unhappy|dissatisf|frustrat|angry", na=False
    )
    positive_mask = df["Customer_Response"].astype(str).str.lower().str.contains(
        "positive|happy|satisf|pleased", na=False
    )
    neutral_mask  = df["Customer_Response"].astype(str).str.lower().str.startswith("neutral", na=False)
 
    df["Customer_Response"] = "Not Discussed"
    df.loc[neutral_mask,  "Customer_Response"] = "Neutral"
    df.loc[positive_mask, "Customer_Response"] = "Positive"
    df.loc[negative_mask, "Customer_Response"] = "Negative"
```

```python
if "Price_Range_Mentioned" in df.columns:

    # Step 1: Normalize text
    df["Price_Range_Mentioned"] = (
        df["Price_Range_Mentioned"]
        .astype(str)
        .str.lower()
        .str.strip()
    )

    # Step 2: Replace noise / non-informative values → NaN
    noise_values = ["not discussed", "not mentioned", "nan", "xxxx", "unavailable", ""]
    df["Price_Range_Mentioned"] = df["Price_Range_Mentioned"].replace(noise_values, "Not Discussed")

    # Step 3: Remove currency symbols and commas
    df["Price_Range_Mentioned"] = df["Price_Range_Mentioned"].str.replace(r"[£$,]", "", regex=True)

    # Step 4: Standardize range format (handle "to", "-", etc.)
    df["Price_Range_Mentioned"] = df["Price_Range_Mentioned"].str.replace(r"\s*-\s*", " to ", regex=True)
    df["Price_Range_Mentioned"] = df["Price_Range_Mentioned"].str.replace(r"\s*to\s*", " to ", regex=True)

    # Step 5: Clean extra spaces again
    df["Price_Range_Mentioned"] = df["Price_Range_Mentioned"].str.strip()
```

### CLEAN CATEGORY COLUMNS

```python
type3_cols = [
    "Churn_Category",
    "Complaint_Category",
    "Customer_Reaction_Category",
    "Agent_Renewal_Pitch_Category",
    "Customer_Renewal_Response_Category",
    "Agent_Response_Category",
    "Justification_Category",
    "Reason_For_Renewal_Category",
    "Agent_Response_To_Cancel_Category",
    "Argument_That_Convinced_Customer_to_Stay_Category",
]

for col in type3_cols:
    if col not in df.columns:
        continue

    # Strip spaces
    df[col] = df[col].astype(str).str.strip()

    # Replace underscores with space
    df[col] = df[col].str.replace("_", " ", regex=False)

    # Remove extra spaces
    df[col] = df[col].str.replace(r"\s+", " ", regex=True)

    # Replace common missing/noise values (case-sensitive handling)
    df[col] = df[col].replace(
        ["nan", "NaN", "N/A", "n/a", "NA", "na", "Not Applicable", ""],
        "Not Mentioned"
    )

    # Fill actual NaN
    df[col] = df[col].fillna("Not Mentioned")
```

#### CLEAN COMPETITOR COLUMNS ACCORDING TO VALUES

```python
if "Competitor_Value_Comparison" in df.columns:

    df["Competitor_Value_Comparison"] = (
        df["Competitor_Value_Comparison"]
        .astype(str)
        .str.lower()
        .str.strip()
        .str.replace(r"[\[\]]", "", regex=True)
    )

    noise_patterns = [
        "not applicable", "n/a", "not mentioned", "not discussed",
        "nan", "xxxx", '"', ""
    ]

    df["Competitor_Value_Comparison"] = np.where(
        df["Competitor_Value_Comparison"].str.contains("better|cheaper|lower|less cost|cost effective|affordable", na=False), "Better Value",
        np.where(
            df["Competitor_Value_Comparison"].str.contains("similar|same|equal", na=False), "Similar Value",
            np.where(
                df["Competitor_Value_Comparison"].str.contains(r"\d|price|vat|£|\$"),
                "Better Value",
                "No Comparison"
            )
        )
    )
```

```python
if "Competitor_Benefits_Mentioned" in df.columns:

    col = (
        df["Competitor_Benefits_Mentioned"]
        .astype(str)
        .str.lower()
        .str.strip()
        .str.replace(r"[\[\]]", "", regex=True)
    )

    noise_patterns = [
        "not applicable", "not mentioned", "not discussed",
        "no benefits", "no mention", "nan", "n/a", ""
    ]

    df["Competitor_Benefits_Mentioned"] = np.where(
        col.str.contains("price|cost|cheap|cheaper|discount|lower", na=False), "Price",
        np.where(
            col.str.contains("service|support|consult|help", na=False), "Service",
            np.where(
                col.str.contains("offering|solution|accreditation|system|process", na=False), "Offering",
                np.where(
                    col.str.contains("easy|ease|simple|simpler|interface", na=False), "Ease",
                    "None"
                )
            )
        )
    )
```
